```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from time import sleep
import requests as rq
import time
from selenium import webdriver
```

```python
page = list(range(1, 5))
url_list = []
url_list.append("https://www.radiookapi.net/politique")

for i in page:
        ii = "https://www.radiookapi.net/politique?page="+str(i)
        url_list.append(ii)
url_list
```

```python
soup_list =[]
for url in url_list:
    soup = rq.get(url)
    soup = bs(soup.content, 'html.parser')
    soup_list.append(soup)
```

```python
news_article = []
for soup in soup_list:
    news_article_divs = soup.find_all('div', attrs={'class':'views-field views-field-title'})
    news_article.append(news_article_divs)
news_article
```

```python
links = []
for i in news_article:
    for ii in i:
        ii = "https://www.radiookapi.net/"+ii.find('a')['href']
        links.append(ii)
links
```

```python
##### title
title = []
for i in news_article:
    for ii in i:
        ii = ii.text.lstrip().rstrip()
        title.append(ii)
title
```

```python
### time 
time_div = []
for soup in soup_list:
    i = soup.find_all('div', attrs={'class':'views-field views-field-created'})
    time_div.append(i)
time_div

time = []
for i in time_div:
    for ii in i:
        iii = ii.text.lstrip().rstrip()
        time.append(iii)
time
```

```python
## news tags
tag_span = []
for soup in soup_list:
    i = soup.find_all('span', attrs={'class':'views-field views-field-field-tags'})
    tag_span.append(i)
tag_span

tags = []
for i in tag_span:
    for ii in i:
        iii = ii.text.lstrip().rstrip()
        tags.append(iii)
tags
```

```python
####### full text
content = []
for i in links:
    soup = rq.get(i)
    soup = bs(soup.content, 'html.parser')
    content.append(soup)

contents = []
for i in content:
    ii = i.find('div', attrs={'class':'field field-name-body field-type-text-with-summary field-label-hidden'})
    iii = ii.text.replace("\n", "").lstrip()
    contents.append(iii)
contents
```

```python
links_df = pd.DataFrame(links)
title_df = pd.DataFrame(title)
tag_df = pd.DataFrame(tags)
time_df = pd.DataFrame(time)
contents_df = pd.DataFrame(contents)


df = pd.concat([links_df,title_df,tag_df,time_df,contents_df], axis=1)
df.columns = ['Link', 'Title', 'Tags', 'Time','Text']
df
```
